```python
import pandas as pd

# 数据路径
data_path = "E:/solid-state patent data/csv_cleaned_ipc.csv"
community_path = "E:/solid-state patent data/新建文件夹/community_partition.csv"
#注意！！！！！！！！！！！！
# 加载数据
df = pd.read_csv(data_path, encoding='utf-8', low_memory=False)
community_df = pd.read_csv(community_path)

# 合并社群信息到专利数据
df = df.merge(community_df, left_on='IPC - 现版', right_on='Node', how='left')
print(df.head())
```

根据社群划分结果，为每个社群提取对应的专利文本（如标题和摘要）。

```python
from collections import defaultdict

# 提取每个社群的专利文本
community_texts = defaultdict(list)

for _, row in df.iterrows():
    community = row['Community']
    text = f"{row['标题 (英语)']} {row['摘要 (英语)']}"  # 合并标题和摘要
    if isinstance(text, str):  # 确保文本不为空
        community_texts[community].append(text)

# 打印每个社群的文本数量
for community, texts in community_texts.items():
    print(f"社群 {community}: {len(texts)} 条文本")
```

加上保存到表格的部分

```python
from bertopic import BERTopic
import os
# 遍历每个社群，进行主题挖掘
community_topics = {}
for community, texts in community_texts.items():
    print(f"正在处理社群 {community}...")
    
    # 初始化 BERTopic 模型
    topic_model = BERTopic(language="english", nr_topics="auto")  # 自动确定主题数量
    
    # 训练模型
    topics, probs = topic_model.fit_transform(texts)
    
    # 保存主题信息
    topic_info = topic_model.get_topic_info()
    topic_info.to_csv(os.path.join(output_folder, f"community_{community}_topics.csv"), index=False, encoding='utf-8')
    
    # 可视化主题分布
    fig = topic_model.visualize_topics()
    fig.write_html(os.path.join(output_folder, f"community_{community}_topics.html"))
    
    # 获取主题关键词并保存为表格
    topics_dict = topic_model.get_topics()
    topic_keywords = []
    for topic_id, keywords in topics_dict.items():
        for word, score in keywords:
            topic_keywords.append({"Topic": topic_id, "Keyword": word, "Score": score})
    
    # 转换为主题关键词表格
    topic_keywords_df = pd.DataFrame(topic_keywords)
    topic_keywords_df.to_csv(os.path.join(output_folder, f"community_{community}_topic_keywords.csv"), index=False, encoding='utf-8')
    
    # 存储主题模型和关键词
    community_topics[community] = {
        "model": topic_model,
        "topics": topics_dict
    }
    
    print(f"社群 {community} 的主题挖掘完成，结果已保存。")
```

2. 生成环形词云嵌套图
为 BERTopic 结果。我们使用 matplotlib 和 wordcloud 库来生成词云。

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import os

def plot_bertopic_wordcloud(bertopic_keywords, output_path):
    # 创建子图
    fig, ax = plt.subplots(figsize=(8, 8))
    
    # 内环：BERTopic 主题词
    inner_wordcloud = WordCloud(
        background_color="white", 
        width=600, 
        height=600, 
        colormap="viridis"  # 可选颜色映射
    ).generate(" ".join(bertopic_keywords))
    
    # 绘制词云
    ax.imshow(inner_wordcloud, interpolation="bilinear")
    ax.axis("off")  # 关闭坐标轴
    
    # 保存图表
    plt.savefig(output_path, dpi=300)
    plt.close()

# 示例：为每个社群生成 BERTopic 环形词云
output_folder = "E:/solid-state patent data/新建文件夹"

for community, data in community_topics.items():
    # 提取 BERTopic 主题关键词
    bertopic_keywords = [word for topic in data["topics"].values() for word, _ in topic]
    
    # 定义输出路径
    output_path = os.path.join(output_folder, f"community_{community}_bertopic_wordcloud.png")
    
    # 生成并保存词云
    plot_bertopic_wordcloud(bertopic_keywords, output_path)
    print(f"BERTopic 词云已保存到: {output_path}")
```

3. 主题演化桑基图
从主题关键词表中提取主题链接
假设每个社群的主题关键词已经保存为 CSV 文件（如 community_{community}_topic_keywords.csv），我们可以读取这些文件并构建跨社群的主题链接。（减少BERTopic步骤重复）

```python
import os
import pandas as pd
from collections import defaultdict

# 定义输入路径
input_folder = "E:/solid-state patent data/新建文件夹"
output_folder = "E:/solid-state patent data/新建文件夹"

# 构建主题链接
def build_topic_links(input_folder):
    topic_links = defaultdict(int)
    
    # 遍历所有社群的主题关键词表
    for file_name in os.listdir(input_folder):
        if file_name.startswith("community_") and file_name.endswith("_topic_keywords.csv"):
            community_id = file_name.split("_")[1]  # 提取社群 ID
            file_path = os.path.join(input_folder, file_name)
            
            # 加载主题关键词表
            df = pd.read_csv(file_path)
            
            # 构建主题名称
            df['Topic_Name'] = df.apply(lambda row: f"Community_{community_id}_Topic_{row['Topic']}", axis=1)
            
            # 统计每个主题的关键词频率
            for _, row in df.iterrows():
                topic_name = row['Topic_Name']
                keyword = row['Keyword']
                
                # 假设关键词在不同社群之间共享，则构建主题链接
                for other_file_name in os.listdir(input_folder):
                    if other_file_name != file_name and other_file_name.endswith("_topic_keywords.csv"):
                        other_community_id = other_file_name.split("_")[1]
                        other_file_path = os.path.join(input_folder, other_file_name)
                        
                        # 加载另一个社群的主题关键词表
                        other_df = pd.read_csv(other_file_path)
                        other_df['Topic_Name'] = other_df.apply(lambda row: f"Community_{other_community_id}_Topic_{row['Topic']}", axis=1)
                        
                        # 如果关键词在另一个社群中出现，则增加链接权重
                        matching_topics = other_df[other_df['Keyword'] == keyword]['Topic_Name']
                        for other_topic_name in matching_topics:
                            topic_links[(topic_name, other_topic_name)] += 1
    
    return topic_links

# 构建主题链接
topic_links = build_topic_links(input_folder)

# 打印主题链接
print("主题链接:")
for link, value in topic_links.items():
    print(f"{link}: {value}")
```

使用主题链接生成桑基图
使用上述提取的主题链接生成桑基图，并保存为 HTML 文件。

```python
import plotly.graph_objects as go

def plot_sankey_diagram(topic_links, output_path):
    # 构建桑基图数据
    labels = list(set([item for pair in topic_links.keys() for item in pair]))
    source_indices = [labels.index(pair[0]) for pair in topic_links.keys()]
    target_indices = [labels.index(pair[1]) for pair in topic_links.keys()]
    values = list(topic_links.values())
    
    # 创建桑基图
    fig = go.Figure(data=[go.Sankey(
        node=dict(
            pad=15,
            thickness=20,
            line=dict(color="black", width=0.5),
            label=labels,
            color="blue"
        ),
        link=dict(
            source=source_indices,
            target=target_indices,
            value=values
        )
    )])
    
    fig.update_layout(title_text="跨社群技术主题关联性", font_size=10)
    
    # 保存为 HTML 文件
    fig.write_html(output_path)
    print(f"桑基图已保存到: {output_path}")

# 定义输出路径
output_path = os.path.join(output_folder, "sankey_diagram.html")

# 生成桑基图
plot_sankey_diagram(topic_links, output_path)
```

太大了打不开，改用网络图可视化

```python
import networkx as nx
import matplotlib.pyplot as plt

def plot_network_graph(topic_links, output_path):
    # 创建网络图
    G = nx.DiGraph()
    
    # 添加节点和边
    for (source, target), weight in topic_links.items():
        G.add_edge(source, target, weight=weight)
    
    # 绘制网络图
    plt.figure(figsize=(12, 8))
    pos = nx.spring_layout(G, k=0.5)
    nx.draw_networkx_nodes(G, pos, node_size=50, node_color="skyblue")
    nx.draw_networkx_edges(G, pos, alpha=0.5, edge_color="gray")
    nx.draw_networkx_labels(G, pos, font_size=8, font_color="black")
    plt.title("跨社群技术主题关联性")
    plt.axis("off")
    
    # 保存为图片文件
    plt.savefig(output_path, dpi=300)
    plt.close()
    print(f"网络图已保存到: {output_path}")

# 示例：生成网络图
network_output_path = os.path.join(output_folder, "topic_links_network.png")
plot_network_graph(topic_links, network_output_path)
```

4. 语义相似度热力图

```python
import os
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt

# 定义输入和输出路径
input_folder = "E:/solid-state patent data/新建文件夹"
output_folder = "E:/solid-state patent data/新建文件夹"

# 加载主题向量
def load_topic_vectors(input_folder):
    community_topics = {}
    
    # 遍历所有社群的主题关键词表
    for file_name in os.listdir(input_folder):
        if file_name.startswith("community_") and file_name.endswith("_topic_keywords.csv"):
            community_id = file_name.split("_")[1]  # 提取社群 ID
            file_path = os.path.join(input_folder, file_name)
            
            # 加载主题关键词表
            df = pd.read_csv(file_path)
            
            # 数据清理：确保 Keyword 是字符串，Score 是数值
            df['Keyword'] = df['Keyword'].astype(str)  # 转换为字符串
            df['Score'] = pd.to_numeric(df['Score'], errors='coerce')  # 转换为数值，无法转换的设为 NaN
            df = df.dropna(subset=['Score'])  # 删除 Score 为 NaN 的行
            
            # 构建主题向量
            topics = {}
            for topic_id, group in df.groupby("Topic"):
                keywords = group["Keyword"].tolist()
                scores = group["Score"].tolist()
                
                # 将关键词和分数作为主题向量
                topics[topic_id] = (keywords, scores)
            
            community_topics[community_id] = topics
    
    return community_topics

# 计算主题向量的嵌入表示（例如通过词向量平均）
def compute_topic_embeddings(topics, embedding_model):
    topic_vectors = {}
    for topic_id, (keywords, scores) in topics.items():
        # 过滤掉无效的关键词和分数
        valid_keywords = []
        valid_scores = []
        for word, score in zip(keywords, scores):
            if isinstance(word, str) and isinstance(score, (float, int)):  # 确保关键词是字符串，分数是数值
                valid_keywords.append(word)
                valid_scores.append(score)
        
        if not valid_keywords:  # 如果没有有效的关键词，跳过该主题
            print(f"警告：主题 {topic_id} 没有有效的关键词，跳过计算。")
            continue
        
        # 使用嵌入模型计算每个主题的向量
        vectors = [embedding_model.encode(word) * score for word, score in zip(valid_keywords, valid_scores)]
        topic_vector = np.mean(vectors, axis=0)  # 取加权平均
        topic_vectors[topic_id] = topic_vector
    
    return topic_vectors

# 示例：加载预训练的嵌入模型
from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# 加载主题关键词并计算主题向量
community_topics = load_topic_vectors(input_folder)

# 存储每个社群的主题向量
community_topic_vectors = {}
for community_id, topics in community_topics.items():
    topic_vectors = compute_topic_embeddings(topics, embedding_model)
    community_topic_vectors[community_id] = topic_vectors

# 绘制语义相似度热力图
def plot_similarity_heatmap(topic_vectors, output_path):
    # 确保 topic_vectors 是二维数组
    topic_vectors_array = np.array(list(topic_vectors.values()))
    if topic_vectors_array.ndim == 1:  # 如果是一维数组，重塑为二维数组
        topic_vectors_array = topic_vectors_array.reshape(-1, 1)
    
    # 计算余弦相似度矩阵
    similarity_matrix = cosine_similarity(topic_vectors_array)
    
    # 绘制热力图
    plt.figure(figsize=(10, 8))
    sns.heatmap(
        similarity_matrix,
        annot=False,  # 不在单元格中标注数字
        cmap="coolwarm",  # 使用 coolwarm 配色方案
        xticklabels=topic_vectors.keys(),  # 横坐标标签
        yticklabels=topic_vectors.keys(),  # 纵坐标标签
        cbar=True  # 显示颜色条作为图例
    )
    plt.title("语义相似度热力图")
    plt.savefig(output_path, dpi=300)
    plt.close()

# 示例：为每个社群生成热力图
for community_id, topic_vectors in community_topic_vectors.items():
    output_path = os.path.join(output_folder, f"community_{community_id}_similarity_heatmap.png")
    plot_similarity_heatmap(topic_vectors, output_path)
    print(f"热力图已保存到: {output_path}")
```
